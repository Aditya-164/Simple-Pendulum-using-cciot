```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("feeds.csv")
data = data.iloc[1:]
data = data.drop(columns=['created_at', 'entry_id', 'latitude', 'longitude', 'elevation', 'status'])
data.to_csv("Thingspeak.csv", index=False)
```

```python
TimePeriod = data['field1'].tolist()
print("TimePeriod:", TimePeriod)

Inst_g = data['field2'].tolist()
print("Inst_g:", Inst_g)

Avg_g = data['field3'].tolist()
print("Avg_g:", Avg_g)

LengthsOfPendulum = data['field4'].tolist()
print("LengthsOfPendulum:", LengthsOfPendulum)
```

```python
plt.plot(np.sqrt(LengthsOfPendulum),TimePeriod)
plt.xlabel('Square root of Lengths')
plt.ylabel('Time Period of Oscillation of Simple ')
plt.title('Plot of T vs sqrt(L)')
plt.grid(True)
plt.show()
```

```python
Inst_g_Array = np.array(Inst_g)
indices = np.arange(1, len(Inst_g)+1)
plt.plot(indices, Inst_g_Array)
plt.ylabel('Instantaneous g')
plt.title('Plot of Instantaneous acceleration due to gravity')
plt.grid(True)
plt.show()
```

```python
Avg_g_Array = np.array(Avg_g)
indices = np.arange(1, len(Avg_g)+1)
plt.plot(indices, Avg_g_Array)
plt.ylabel('Avg_g')
plt.title('Plot of Average acceleration due to gravity')
plt.grid(True)
plt.show()
```

```python
plt.plot(LengthsOfPendulum, TimePeriod, label='Time Period')
plt.plot(LengthsOfPendulum, Inst_g, label='Instantaneous g value')
plt.plot(LengthsOfPendulum, Avg_g, label='Average g value')

plt.xlabel('List 3')
plt.ylabel('Values')
plt.title('Plot of List 1 and List 2 vs List 3')
plt.legend()  # Show legend
plt.grid(True)
plt.show()
```
